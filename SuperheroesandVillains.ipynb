Randomized Comparision between One Superhero and One Villain

```python
# records from Yellowjacket II

marvel_df.loc[(marvel_df['Alignment'] == 'good') & (marvel_df['Name'] == 'Yellowjacket II')]
```

```python
# records for Amazo

marvel_df.loc[(marvel_df['Alignment'] == 'bad') & (marvel_df['Name'] == 'Amazo')]
```

```python
# vizualization of stats between Yellowjacket II and Amazo

# create dataframe for Yellowjacket II and Amazo
stats = pd.DataFrame({
    'Intelligence': [50,75],
    'Strength': [10,100],
    'Speed': [35,100],
    'Durability': [28,100],
    'Power': [31,100],
    'Combat': [28,100]},
    index= ['Yellowjacket II', 'Amazo'])
    
stats.head()

```

```python
# create bar graph showing difference in skillset between Yellowjacket II and Amazo

stats.reset_index().plot(
    x='index', y=['Intelligence','Strength','Speed'],
    kind='bar')

plt.title('Yellowjacket II vs Amazo')
plt.xlabel('Characters')
plt.ylabel('Points')
```

```python
stats.reset_index().plot(
    x='index', y=['Durability','Power','Combat'],
    kind='bar')

plt.title('Yellowjacket II vs Amazo')
plt.xlabel('Characters')
plt.ylabel('Points')
```
